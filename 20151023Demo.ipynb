```python
newsary = []
for news in soup.select('.rtddt'):        
    category = news.select('h2')[0].text 
    time = news.select('time')[0].text 
    url = domain + news.select('a')[0]['href']
    main = news.select('h1')[0].text
    m = re.match('(.*)\((\d+)\)',main)
    if m:
        title = m.group(1)
        popularity = int(m.group(2))
    dic = {'title':title, 'category':category, \
           'time':time, 'url':url, \
           'popularity':popularity}
    newsary.append(dic)
```

```python
appleurl = 'http://www.appledaily.com.tw/realtimenews/section/new/{}'
domain = 'http://www.appledaily.com.tw'
newsary = []
for page in range(1,31):
    #print url.format(page)
    res = requests.get(appleurl.format(page))
    #print appleurl.format(page)
    soup = bs(res.text)
    for news in soup.select('.rtddt'):        
        category = news.select('h2')[0].text 
        time = news.select('time')[0].text 
        if 'http://www.appledaily.com.tw' not in news.select('a')[0]['href']:
            url = domain + news.select('a')[0]['href']
        else:
            url = news.select('a')[0]['href']
        main = news.select('h1')[0].text
        m = re.match('(.*)\((\d+)\)',main)
        if m:
            title = m.group(1)
            popularity = int(m.group(2))
        dic = {'title':title, 'category':category, \
               'time':time, 'url':url, \
               'popularity':popularity}
        newsary.append(dic)
```

```python
def bigram(input_sentence):
    sentence  = input_sentence
    ret = []
    for i in range(0, len(sentence) - 2 + 1):
        ret.append(sentence[i:i+2])
    return ret
```

```python
def trigram(input_sentence):
    sentence  = input_sentence
    ret = []
    for i in range(0, len(sentence) - 3 + 1):
        ret.append(sentence[i:i+3])
    return ret
```

```python
dic = {}
for news in newsary:
    for word in bigram(news['title']):
        if word not in dic:
            dic[word] = 1
        else:
            dic[word] = dic[word] + 1
```

```python
def ngram(input_sentence, n = 2):
    word_dic = {}
    sentence  = input_sentence
    for i in range(0, len(sentence) - n + 1):
        if sentence[i:i+n] not in word_dic:
            word_dic[sentence[i:i+n]] = 1
        else:
            word_dic[sentence[i:i+n]] = word_dic[sentence[i:i+n]] + 1
    return word_dic
```

```python
skip_dic = '?.,、　「！】？：《」》～【'.decode('utf-8')
def ngram(input_sentence, n = 2, skip_word= skip_dic):
    word_dic = {}
    sentence  = input_sentence
    for i in range(0, len(sentence) - n + 1):
        invalid = len([word for word in sentence[i:i+n] if word in skip_word])
        if sentence[i:i+n] not in word_dic and invalid ==0:
            word_dic[sentence[i:i+n]] = 1
        elif invalid ==0:
            word_dic[sentence[i:i+n]] = word_dic[sentence[i:i+n]] + 1
    return word_dic
```
