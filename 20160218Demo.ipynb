```python
import jieba
import sqlite3
db = sqlite3.connect('news2.sqlite') 
cur = db.cursor()
cur.execute('select title, summary, category from news_entry')
allNews = cur.fetchall()

corpus = []
tags = []
title = []
for rec in allNews:
    if (rec[2] == '娛樂'.decode('utf-8')) or \
       (rec[2] == '社會'.decode('utf-8')) or \
       (rec[2] == '財經'.decode('utf-8'))  :
        corpus.append(' '.join(jieba.cut(rec[1])))
        tags.append(rec[2]) 
cur.close()
db.close()
```

```python
import numpy as np
import math
def entropy_compute(classifier):
    elist = []
    for rec in zip(*np.vstack(classifier.feature_count_)):
        e = 0
        s = sum(rec)
        for ele in rec:
            if ele > 0:
                e = e - (ele / s ) * math.log(ele/s,2)
        elist.append(e) 
    return elist
```

```python
def most_informative_feature_for_class(vectorizer, classifier, classlabel, n=10):
    labelid = list(classifier.classes_).index(classlabel)
    feature_names = vectorizer.get_feature_names()
    #print classifier.feature_count_[-0]
    count_sum =  [sum(rec) for rec in zip(*np.vstack(classifier.feature_count_))]
    topn = sorted(zip(classifier.coef_[labelid], feature_names, entropy_compute(classifier), count_sum))[-n:]
    return topn
    #for coef, feat, ent, csum in topn:
    #    print classlabel, feat, coef, ent, csum
```
